```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_columns', None, 'display.max_rows', None)

# Data_Set_EM

df = pd.read_csv('../../data/data_imputed_v2.csv', index_col=0)

PATH = '../../Model Save/Data_Set_EM/'
# PATH = '/content/drive/MyDrive/Project/Model Save/'

# df.info(verbose=True)
```

```python
df.head()
```

```python
df['franchise_dealer'] = df['franchise_dealer'].astype('int')
df['is_new'] = df['is_new'].astype('int')

# df.head()
```

```python
df.info(verbose=True)
```

```python
y = df.loc[:,'price']
X= df.loc[:,df.columns != 'price']

X.head()
X.tail()
```

```python
X.shape
```

```python
import math
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import pickle
import os
import joblib

import warnings 
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

#### Split train and test data

```python
X_train = X.iloc[:15000,:]
print(X_train.shape)

X_test = X.iloc[15000:,:]
print(X_test.shape)

y_train = y.iloc[:15000]
print(y_train.shape)

y_test = y.iloc[15000:]
print(y_test.shape)
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Single Base Model

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

slr = LinearRegression()

slr.fit(X_train, y_train)
```

```python
joblib.dump(slr, PATH+'lrclassifier_141.pkl')
loaded_slr = joblib.load(PATH+'lrclassifier_141.pkl')
```

```python

y_train_slr = loaded_slr.predict(X_train)

y_test_slr = loaded_slr.predict(X_test)
```

```python
LR_MAE_train  = mean_absolute_error(y_train, y_train_slr)
LR_MSE_train = mean_squared_error(y_train, y_train_slr)
LR_RMSE_train = math.sqrt(mean_squared_error(y_train, y_train_slr))
LR_Rsquare_train = r2_score(y_train, y_train_slr)

LR_MAE_test = mean_absolute_error(y_test, y_test_slr)
LR_MSE_test = mean_squared_error(y_test, y_test_slr)
LR_RMSE_test = math.sqrt(mean_squared_error(y_test, y_test_slr))
LR_Rsquare_test = r2_score(y_test, y_test_slr)
```

```python
print(f'LR MAE train: {LR_MAE_train:.3f}')
print(f'LR MSE train: {LR_MSE_train:.3f}')
print(f'LR RMSE train: {LR_RMSE_train:.3f}')
print(f'LR R^2 train: {LR_Rsquare_train:.3f}')

print(f'LR MAE test: {LR_MAE_test:.3f}')
print(f'LR MSE test: {LR_MSE_test:.3f}')
print(f'LR RMSE test:{LR_RMSE_test:.3f}')
print(f'LR R^2 test: {LR_Rsquare_test:.3f}')
```

#### Lasso linear model

##### Grid Search

- `Lasso`(Least Absolute Shrinkage and Selection Operator) with L1 regularization
